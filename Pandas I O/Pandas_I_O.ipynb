```python
url2 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/pacientes_2.json'
```

```python
pacientes_2 = pd.read_json(url2)
pacientes_2.head()
```

```python
#salvando arquivo json
df_normalizado.to_json('pacientes_normalizados.json')
pd.read_json('/content/pacientes_normalizados.json')
```

#Banco de dados

```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
import pandas as pd
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
dados.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

```python
empregados.to_sql('empregados', con=engine, index=False)
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente','Grau_escolaridade', 'Rendimento_anual'])
```
