##Compreendendo os dados

```python
#aplicando normalização

dados= pd.json_normalize(dados['info_moveis'])
```

```python
dados.head()
```

```python
#Muitos dados estão agrupados em listas, iremos usar o método explode para ajustar isso
colunas = list(dados.columns)
```

```python
colunas
```

```python
dados = dados.explode(colunas[3:])
dados
```

```python
dados.reset_index(inplace=True, drop=True)
```

```python
dados
```

```python
dados.info()
```

```python
#object = vários tipos de dados diferentes , não especificado ou somente string
```

##Convertendo dados numéricos

```python
import numpy as np
```

```python
dados.head()
```

```python
dados['max_hospedes']=dados['max_hospedes'].astype(np.int64)
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
dados.info()
```

```python
dados['avaliacao_geral']= dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

##Coluna preço com $

```python
dados['preco']
```

```python
#utilizaremos apply com a função lambda
dados['preco']=dados['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados.head()
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
dados.info()
```

##Transformação de Múltiplas Colunas

```python
dados[['taxa_deposito','taxa_limpeza']]=dados[['taxa_deposito','taxa_limpeza']].map(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados[['taxa_deposito','taxa_limpeza']]=dados[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

##Tokenização

```python
dados.head()
```

```python
dados['descricao_local']= dados['descricao_local'].str.lower()
dados.head()
```

##Removendo Caracteres com regex

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local']=dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
```

```python
dados['descricao_local']=dados['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ', regex=True)
```

```python
dados.head()
```

```python
#Tokenização em strings com split

dados['descricao_local']=dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades']=dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
dados.head()
```

```python
dados.tail()
```

```python
dados['comodidades']=dados['comodidades'].str.split(',')
```

```python
dados.head()
```

#Manipulação de Strings

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/banco_clientes.csv')
```

```python
df.head()
```

```python
#Padronizando nomes = letra maiúscula
df['nome'] = df['nome'].str.upper()
```

```python
df.head()
```

```python
#Padronizando a coluna telefone
df['telefone'] = df['telefone'].str.replace('(','', regex=False).str.replace(')','', regex = False)
```

```python
df.head()
```

```python
#Padronizando ceps
df['cep']=df['endereco'].str.extract('(\d{5}-\d{3})')
```

```python
df
```

```python
df['endereco']=df['endereco'].str.replace('(\d{5}-\d{3})','', regex=True).str.strip()
```

```python
df
```

```python
#coluna para rua e outra para numero do end
df['endereco'].str.split('\s(?=\d+$)',expand=True)
```

```python
#coluna para rua e outra para numero do end
df['endereco'].str.replace(',','').str.split('\s(?=\d+$)',expand=True)
```

```python
df[['rua','número_casa']] = df['endereco'].str.replace(',','').str.split('\s(?=\d+$)',regex=True,expand=True)
```

```python
df.head()
```

```python
#drop endereço
df.drop(['endereco'],axis=1)
df
```
